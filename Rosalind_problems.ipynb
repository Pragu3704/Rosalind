```python
#Translate mRNA to protein

from google.colab import files
upload = files.upload()                            #Uplaoding the data

name_of_file=list(upload.keys())[0]                 #Getting file name

with open(name_of_file,'r') as file:
  mRNA_strand=file.read()                           #Open file in read mode, reading the data textfile and converting it to a string

from Bio import Seq
protein = Seq.translate(mRNA_strand, to_stop="True")  #Translate method looks for codons and returns equivalent amino acid, to_stop="True" stops the translation of the protein at stop codon
print(protein)

```
